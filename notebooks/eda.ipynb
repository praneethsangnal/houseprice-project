plotting the target salesprice 


```python
plt.figure(figsize=(8,5))
sns.histplot(df["SalePrice"], kde=True)
plt.title("SalePrice Distribution")
plt.show()
```

top 10 most influential numeric features


```python
corr = df.corr(numeric_only=True)["SalePrice"]
corr.sort_values(ascending=False).head(10)
```

correlation heatmap

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(numeric_only=True),
            cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
print(df["SalePrice"].skew())
```

highly right skewed so applying log function for symmetricity in bell curve


```python
plt.figure(figsize=(8,5))
sns.histplot(np.log1p(df["SalePrice"]), kde=True)
plt.title("SalePrice Distribution")
plt.show()
```
